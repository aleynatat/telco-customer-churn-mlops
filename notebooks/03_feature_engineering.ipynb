```python
from operator import index

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/Telco-Customer-Churn-Cleaned.csv')
df.head()
```

```python
df['is_high_risk_billing'] = np.where((df['PaymentMethod'] == "Electronic check") & (df['Contract'] == "Month-to-month"), 1, 0)
df['is_high_risk_billing'].head()
df.head()
```

```python
services = ['PhoneService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

df['Total_Services'] = (df[services] == 'Yes').sum(axis=1)

```

```python
columns = df.select_dtypes(include=['object', 'str']).columns
print(columns)
for c in columns:
    print(c)
    print(df[c].unique())
    print("-----------------------------------")
```

```python
binary_columns= ['gender', 'Partner', 'Dependents', 'PhoneService', 'OnlineSecurity', 'OnlineBackup','DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling']

nominal_columns= ['MultipleLines', 'InternetService', 'Contract', 'PaymentMethod', 'tenure_year' ]
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

encoder = OneHotEncoder(drop= 'first', sparse_output=False)

encoded_data = encoder.fit_transform(df[binary_columns])

encoded_df = pd.DataFrame(
    encoded_data,
    columns=encoder.get_feature_names_out(binary_columns),
    index= df.index,
)

final_df = pd.concat(
    [df.drop(columns=binary_columns), encoded_df],
    axis=1
)

print("\nOne-Hot Encoded Data:")
print(final_df)

final_df.head()
```

```python
final_df['Churn'] = np.where(df['Churn'] == "Yes", 1, 0)
final_df.head()
```

```python
encoder2 = OrdinalEncoder()

encoded_data = encoder2.fit_transform(final_df[nominal_columns])

encoded_df = pd.DataFrame(
    encoded_data,
    columns=encoder2.get_feature_names_out(nominal_columns),
)

df = pd.concat(
    [final_df.drop(columns=nominal_columns), encoded_df],
    axis=1
)

print("\nOrdinal Encoded Data:")
print(df)
```

```python
df.head()
```

```python
df.drop('customerID', axis=1, inplace=True)
df.head()
```
